```python
#function with arguments
def cal(a,b,o):
    if o == "+":
        print(a+b)
    elif o == "-":
        print(a-b)
    elif o == "*":
        print(a*b)
    elif o == "/":
        print(a/b)
    else:
        print("Invalid operator")
        
        
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + - * /  ")
     
cal(a,b,o)
```

```python
#function with arguments and return type
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b

def cal(a,b,o):
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("Invalid operator")
        
        
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + - * /  ")
     
cal(a,b,o)
```

```python
def info(l):
    name,age,add = l
    data = f"Hello I am {name}. I am from {age}. I am {add}"
    return data

name = input("Enter name = ")
age = int(input("Enter age = "))
add = input("Enter address = ")
l = [name,age,add]
print(info(l))
```

```python
def info(l):
    data = f"Hello I am {l['Name']}. I am from {l['Add']}. I am {l['Age']}"
    return data

name = input("Enter name = ")
age = int(input("Enter age = "))
add = input("Enter address = ")
l = {"Name":name,"Age":age,"Add":add}
print(info(l))
```
