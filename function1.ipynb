```python
#function with argument and return type
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b

def cal(a,b,o):
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("Invalid operator")
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + - * /  ")
cal(a,b,o)
```
